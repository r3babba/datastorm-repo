```python
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', 1000)
from sklearn import metrics
import matplotlib.pyplot as plt
import seaborn as sns
import prince
```

```python
#importing data
df = pd.read_csv('TrainData.csv')

#making copy to work with
df_copy = df.copy()

X = df_copy.drop(['NEXT_MONTH_DEFAULT','Client_ID','Pay_Time','NO_PAY_DEC'], axis=1)
y = df_copy["NEXT_MONTH_DEFAULT"]
```

```python
#MFA Analysis

#groupng responses for the MFA model
groups = {
  "demos": ['Balance_Limit_V1','Gender','EDUCATION_STATUS','MARITAL_STATUS','AGE'],
  "pay_time": ['PAY_JULY','PAY_AUG','PAY_SEP','PAY_OCT','PAY_NOV','PAY_DEC'],
  "due_amt": ['DUE_AMT_JULY','DUE_AMT_AUG','DUE_AMT_SEP','DUE_AMT_OCT','DUE_AMT_NOV','DUE_AMT_DEC'],
  "paid_amt": ['PAID_AMT_JULY','PAID_AMT_AUG','PAID_AMT_SEP','PAID_AMT_OCT','PAID_AMT_NOV','PAID_AMT_DEC']
}

mfa = prince.MFA(groups=groups,n_components=2,n_iter=3,copy=True,check_input=True,engine='auto',random_state=42)
mfa = mfa.fit(X)
mfa_loadings = mfa.row_coordinates(X)

loadings = mfa.row_coordinates(X)
loadings['Cluster'] = y 

plt.figure(figsize=(15,15))
sns.scatterplot(x=0, y=1, data=loadings, hue="Cluster", palette="Set1", legend='full', s=10)
```

```python
#Z Score from MFA

from scipy import stats

z = np.abs(stats.zscore(mfa_loadings))
threshold = 3
outliers = np.where(z > 3)
outliers
```

```python
#Dropping Outliers
clean_df = df_copy.drop(outliers[0])
print(df_copy.shape)
print(clean_df.shape)
```

```python
#Saving New Data 
filename = 'clean_df.csv'
clean_df.to_csv(filename,index=False)

print('Saved file: ' + filename)
```
